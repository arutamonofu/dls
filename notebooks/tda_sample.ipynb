```python
DATA_ROOT = '../data'
dataset = load_qm9(root=DATA_ROOT)
```

Выберем молекулу с индексом, например, `100`.

```python
molecule_idx = 100
sample_molecule = dataset[molecule_idx]
```

Фильтрация Вьеториса-Рипса для построения персистентной диаграммы.

- Связные компоненты;
- Циклы.

```python
tda_engine = TDAEngine(homology_dims=(0, 1))
pos_numpy = sample_molecule.pos.numpy()[None, :, :]

diagram = tda_engine.vr.fit_transform(pos_numpy)

fig = plot_diagram(diagram[0])
fig.update_layout(title=f'Персистентная диаграмма для молекулы {molecule_idx}')
fig.show()
```

Персистентная энтропия:

1. Время жизни топологической единицы: $L_i = Death_i - Birth_i$.
2. Сложение жизней всех единиц.
3. Определение доли каждой единицы в общей жизни: $p_i = L_i / \sum L_j$.
4. Энтропия: $E = -\sum p_i \log(p_i)$.
